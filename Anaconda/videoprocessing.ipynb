```python
import cv2,time
```

```python
video=cv2.VideoCapture(0)
```

```python
a=0
while True:
    a=a+1
    check,frame,=video.read(1)
    
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('Capturing',gray)
    
    key=cv2.waitKey(1)
    print(gray)
    
    if key == ord('q'):
        break
        
print(a)
video.release()
cv2.distroyAllWindows
```
