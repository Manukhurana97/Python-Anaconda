```python
import requests
from bs4 import BeautifulSoup

r=requests.get('https://www.magicbricks.com/property-for-sale/residential-real-estate?bedroom=4&proptype=Residential-House,Villa&cityName=New-Delhi')
c=r.content

soup=BeautifulSoup(c,'html.parser')
#print(soup.prettify())

all=soup.find_all('div',{'class':'flex relative clearfix m-srp-card__container'})
all

all[0].find('div',{'class':'m-srp-card__price'}).text


all[0].find('div',{'class':'m-srp-card__summary__info'}).text.replace('\xa0','')

all[0].find('span',{'class':"m-srp-card__title__bhk"}).text.replace('\t','').replace('\n','').replace(' ','')


# all
```

```python
l=[]
for item in all:
    d={}
    d['Price']=item.find('div',{'class':'m-srp-card__price'}).text.replace('â‚¹','')
    d['BHK']=item.find('span',{'class':"m-srp-card__title__bhk"}).text.replace('\t','').replace('\n','').replace(' ','')
    
    for col_group in item.find_all('div',{'class':'m-srp-card__summary js-collapse__content'}):
        for summ_title,summ_info in zip(col_group.find_all('div',{'class':'m-srp-card__summary__title'}),col_group.find_all('div',{'class':'m-srp-card__summary__info'})):
#             print(summ_title.text," ",summ_info.text)
            if 'floor' in summ_title.text:
                d['Floor']=summ_info.text
            if 'facing' in summ_title.text:
                d['Facing']=summ_info.text
            if 'car parking' in summ_title.text:
                d['Car Parking']=summ_info.text
            if 'ownership' in summ_title.text:
                d['OWNERSHIP']=summ_info.text    
                

    d['Description']=item.find('div',{'class':'m-srp-card__description js-content-read-more truncated'}).text.replace('read more','').replace('\t','').replace('\n','').replace('   ','')             
    
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```
